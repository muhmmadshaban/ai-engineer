```python
from collections import deque
class Queue:
    def __init__(self):
        self.buffer=deque()
    def enqueue(self,value):
        self.buffer.appendleft(value)
    def dequeue(self):
        return self.buffer.pop()
    def length(self):
        return len(self.buffer)
    def isempty(self):
        return len(self.buffer)==0
    
    
```

```python
q=Queue()
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)
```

```python
q.dequeue()
```

```python
q.length()
```
