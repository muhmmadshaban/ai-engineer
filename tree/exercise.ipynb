```python
class Treenode():
    def __init__(self,data,designation=None):
        self.data = data
        self.designation=designation
        self.children=[]
        self.parent=None
    def add_child(self,child):
        child.parent=self
        self.children.append(child)
        return 
    def get_level(self):
        level=0
        p=self.parent
        while p:
            level+=1
            p=p.parent
        return level
    

    def print_tree(self,level,option=None):
        space =' '*self.get_level()*3
        prefix=space+ '|__ ' if self.parent else ""


        I=0
        if option=="designation":
            print(prefix+ self.designation)
        elif option=="both":
            print(prefix+ self.data +"("+ self.designation+")")

        else:
            if I<=level:
                print(prefix+ self.data)
                I=I+1    
        for x,e in enumerate(self.children):
            e.print_tree(level-1,option)

```

```python

def build_management():
    ceo= Treenode("Nilpul","CEO")
    cto=Treenode("Chinmay","CTO")
    ih=Treenode("Vishvkarma","Infrastructure head")
    cm=Treenode("Dhaval","Cloud Manager")
    ad=Treenode("Abhijeet","App Devloper")
    ih.add_child(cm)
    ih.add_child(ad)
    cto.add_child(ih)
    app=Treenode("Aamir","Application Head")
    cto.add_child(app)

    hr=Treenode("Gels","HR HEAD")
    rm=Treenode("Petel","Requirement Manager")
    pm=Treenode("Waqas","Policy Manager")
    hr.add_child(rm)
    hr.add_child(pm)
    ceo.add_child(cto)
    ceo.add_child(hr)
    return ceo




if __name__ == '__main__':
    root_node = build_management()
    root_node.print_tree("name") # prints only name hierarchy
    root_node.print_tree("designation") # prints only designation hierarchy
    root_node.print_tree("both") # prints both (name and designation) hierarchy
```
